Реализовать следующие структуры данных:
  
● <b> Стек (stack): </b>  
Операции для стека: 
<i> инициализация, проверка на пустоту, добавление нового элемента в начало, извлечение элемента из начала;  </i>  


● <b> Дек (двусторонняя очередь, deque):  </b>  
Операции для дека: 
<i> инициализация, проверка на пустоту, добавление нового элемента в начало, добавление нового элемента в конец, извлечение элемента из начала, извлечение элемента из конца. </i>  
  
<i> Разработать программу обработки данных, содержащихся в заранее подготовленном txt-файле, в соответствии с заданиями, применив указанную в задании структуру данных. Результат работы программы вывести на экран и сохранить в отдельном txt-файле. </i>

```python
#стек

class Stack (object):
    
    #инициализация
    def __init__ (self):
        self.list = []
    
    #проверка на пустоту
    def is_empty (self):
        return len(self.list) == 0
    
    #добавление нового элемента в конец
    def push (self, item):
        self.list.append(item)
        
    #извлечение элемента из конца
    def pop (self):
        return self.list.pop()
        
    #просмотр верхнего элемента
    def top (self):
        if not self.is_empty():
            return self.list[-1]
        
    #определение размера
    def length (self):
        return len(self.list)
    
    #вывод содержимого на экран
    def print_content (self):
        s = "stack content: "
        for item in self.list:
            s += str(item) + ', '
        print (s[:len(s)-2])

stack = Stack()
stack.push(12)
stack.push(2)
stack.push(5)
stack.push(13)
stack.print_content()
print(stack.pop())
stack.print_content()
print(stack.top())
print(stack.length())
```

```python
#дек

class Deque (object):
    
    #инициализация
    def __init__ (self):
        self.list = []
    
    #проверка на пустоту
    def is_empty (self):
        return len(self.list) == 0
    
    #добавление нового элемента в конец (в верх)
    def push_back (self, item):
        self.list.append(item)
        
    #извлечение элемента из конца (сверху)
    def pop_back (self):
        return self.list.pop()
        
    #добавление нового элемента в начало (в низ)
    def push_front (self, item):
        self.list.insert(0, item)
        
    #извлечение элемента из начала (снизу)
    def pop_front (self):
        ans = self.list[0]
        del self.list[0]
        return ans
        
    #просмотр верхнего элемента
    def top (self):
        if not self.is_empty():
            return self.list[-1]
    
    #просмотр нижнего элемента
    def back (self):
        if not self.is_empty():
            return self.list[0]
        
    #определение размера
    def length (self):
        return len(self.list)
    
    #вывод содержимого на экран
    def print_content (self):
        s = "deque content: "
        for item in self.list:
            s += str(item) + ', '
        print (s[:len(s)-2])

deq = Deque()
deq.push_back(1)
deq.push_back(2)
deq.push_back(3)
deq.push_front(0)
deq.push_front(-1)
deq.print_content()
print(deq.pop_back())
print(deq.pop_front())
deq.print_content()
print(deq.top())
print(deq.back())
print(deq.length())
```

### Задание №1
Отсортировать строки файла, содержащие названия книг, в алфавитном порядке с использованием двух деков.

```python
deq = Deque()
deq_sorted = Deque()

with open('Lab4_txt/1.txt', encoding='utf-8') as file:
    for i in file:
        i = i.replace("\n", "")
        deq.push_back(i)
file.close()

def answer_to_txt (deq):
    file = open('Lab4_txt/1_ans.txt', 'w', encoding='utf-8')
    while not deq_sorted.is_empty() :
        file.write(str(deq_sorted.pop_front()) + '\n')
    file.close()

deq.print_content()
    
while not deq.is_empty():
    big_boy = ''
    if deq.back() > deq.top():         #1st > last
        big_boy = deq.pop_front()  #last > 1st
    else:
        big_boy = deq.pop_back()

    if deq_sorted.is_empty():      #if empty
        deq_sorted.push_back(big_boy)
    elif (deq_sorted.length() == 1):    #if has 1 element
        if deq_sorted.back() < big_boy:
            deq_sorted.push_back(big_boy)
        else:
            deq_sorted.push_front(big_boy)
    else:                           #if has 2+ elements
        if deq_sorted.back() >= big_boy:            #gogo first
            deq_sorted.push_front(big_boy)
        elif deq_sorted.top() <= big_boy:         #gogo last
            deq_sorted.push_back(big_boy)
        else:                                  #well.. out u go, boys
            big_boys_little_bro = deq_sorted.top()
            while big_boys_little_bro > big_boy:
                big_boys_little_bro = deq_sorted.pop_back()
                deq.push_back(big_boys_little_bro)
                big_boys_little_bro = deq_sorted.top()
            deq_sorted.push_back(big_boy)

print()
deq_sorted.print_content()
answer_to_txt(deq_sorted)
```

### Задание №2
Дек содержит последовательность символов для шифровки сообщений. Дан текстовый файл, содержащий зашифрованное сообщение. Пользуясь деком, расшифровать текст. 
Известно, что при шифровке каждый символ сообщения заменялся следующим за ним в деке по часовой стрелке через один.

```python
deq = Deque()
for i in range(32):  #fill dec
    deq.push_back(chr(ord("а") + i))

def find_letter (c):
    while deq.top() != c:
        deq.push_front(deq.pop_back())
    #here c is on the top and its value is second from the bottom, lets bring the value to the bottom
    deq.push_back(deq.pop_front())

file = open('Lab4_txt/2.txt', encoding='utf-8')
text = ''
prev_text = ''
for line in file:
    for symb in line:
        prev_text += symb
        if symb >= 'а' and symb <= 'я':
            find_letter(symb)
            text += deq.back()
        else:
            text += symb
file.close()      

file = open('Lab4_txt/2_ans.txt', 'w', encoding='utf-8')
file.write(text)
file.close()

print(prev_text)
print(text)
```

### Задание №3
Даны три стержня и n дисков различного размера. Диски можно надевать на стержни, образуя из них башни. Перенести n дисков со стержня А на стержень С, сохранив их первоначальный порядок. При переносе дисков необходимо соблюдать следующие правила:
- на каждом шаге со стержня на стержень переносить только один диск;
- диск нельзя помещать на диск меньшего размера;
- для промежуточного хранения можно использовать стержень В.
Реализовать алгоритм, используя три стека вместо стержней А, В, С. Информация о дисках хранится в исходном файле.

```python
def move(a,b):
    if a.length() == 0 and b.length() != 0:
        a.push(b.pop())
    elif a.length() != 0 and b.length() == 0:
        b.push(a.pop())
    elif a.length() == 0 and b.length() == 0:
        return
    elif a.top() < b.top():
        a.push(b.pop())
    else:
        b.push(a.pop())     

stack_a = Stack()
stack_b = Stack()
stack_c = Stack()
stack_a.push(17)
stack_a.push(24)
stack_a.push(52)
stack_a.push(11)
stack_a.push(3)
n = -1
while n > stack_a.length() or n <= 0:
    n = int (input('Сколько дисков вы хотите перенести? (не больше 5)   '))
    if n > stack_a.length() or n <= 0:
        print ('Такое число не подходит!')
print('A: ')
stack_a.print_content()

if n%2==0:
    while stack_c.length() != n:
        move(stack_a, stack_b)
        move(stack_a, stack_c)
        if stack_c.length() != n:
            move(stack_b, stack_c)
else:
    while stack_c.length() != n:
        move(stack_a, stack_c)
        move(stack_a, stack_b)
        if stack_c.length() != n:
            move(stack_b, stack_c)
print('C: ')
stack_c.print_content()

```

### Задание №4
Дан текстовый файл с программой на алгоритмическом языке. За один просмотр файла проверить баланс круглых скобок в тексте, используя стек.

```python
stack = Stack()
file = open('Lab4_txt/4.txt', encoding='utf-8')

ok_fl = True
for line in file:
    if not ok_fl:
        break
    for sym in line:
        if sym == '(':
            stack.push(sym)
        elif sym == ')':
            if stack.length() != 0:
                stack.pop()
            else:
                ok_fl = False
                break
file.close()

file = open('Lab4_txt/4_ans.txt','w', encoding='utf-8')
if ok_fl and stack.length() == 0:
    print(True)
    file.write("True")
else:
    print(False)
    file.write("False")
file.close()
```

### Задание №5
Дан текстовый файл с программой на алгоритмическом языке. За один просмотр файла проверить баланс квадратных скобок в тексте, используя дек.

```python
dec = Deque()
file = open('Lab4_txt/5.txt', encoding='utf-8')

ok_fl = True
for line in file:
    if not ok_fl:
        break
    for symb in line:
        if symb == '[':
            dec.push_back(symb)
        elif symb == ']':
            if dec.length() != 0:
                dec.pop_back()
            else:
                ok_fl = False
                break
file.close()

file = open('Lab4_txt/5_ans.txt','w', encoding='utf-8')
if ok_fl and dec.length() == 0:
    print(True)
    file.write("True")
else:
    print(False)
    file.write("False")
file.close()
```

### Задание №6
Дан файл из символов. Используя стек, за один просмотр файла напечатать сначала все цифры, затем все буквы, и, наконец, все остальные символы, сохраняя исходный порядок в каждой группе символов.

```python
numbers = Stack()
letters = Stack()
other = Stack()
file = open('Lab4_txt/6.txt', encoding='utf-8')
for line in file:
    for symb in line:
        if symb >= '0' and symb <= '9':
            numbers.push(symb)
        elif symb >= 'a' and symb <= 'z' or symb >= 'A' and symb <= 'Z':
            letters.push(symb)
        else:
            other.push(symb)
file.close()
file = open('Lab4_txt/6_ans.txt','w', encoding='utf-8')
text = ''
while other.length() != 0:
    text = other.pop() + text
while letters.length() != 0:
    text = letters.pop() + text
while numbers.length() != 0:
    text = numbers.pop() + text
print(text)
file.write(text)
file.close()
```

### Задание №7
Дан файл из целых чисел. Используя дек, за один просмотр файла напечатать сначала все отрицательные числа, затем все положительные числа, сохраняя исходный порядок в каждой группе.

```python
dec = Deque()
file = open('Lab4_txt/7.txt', encoding='utf-8')
file_ans = open('Lab4_txt/7_ans.txt','w', encoding='utf-8')
text = ''
for line in file:
    line = line.rstrip()
    num = int(line)
    if num < 0 :  #если отрицательное - печатаем сразу
        print(line)
        file_ans.write(line + '\n')
    else:       #если положительное - отложим на попозже
        dec.push_front(line)
while dec.length() != 0:
    line = dec.pop_back()
    print(line)
    file_ans.write(line + '\n')

file.close()
file_ans.close()
```

### Задание №8
Дан текстовый файл. Используя стек, сформировать новый текстовый файл, содержащий строки исходного файла, записанные в обратном порядке: первая строка становится последней, вторая – предпоследней и т.д.

```python
file = open('Lab4_txt/8.txt', encoding='utf-8')
stack = Stack()
for line in file:
    stack.push(line)
file.close()
file = open('Lab4_txt/8_ans.txt','w', encoding='utf-8')
text = ''
while stack.length() != 0:
    tmp = stack.pop()
    text += tmp
    file.write(tmp)

print(text)
file.close()
```
