### Задание №2
Написать генератор случайных матриц(многомерных), который принимает
опциональные параметры <b>m</b>, <b>n</b>, <b>min_limit</b>, <b>max_limit</b>, где <b>m</b> и <b>n</b> указывают размер
матрицы, а <b>min_lim</b> и <b>max_lim</b> - минимальное и максимальное значение для
генерируемого числа.

```python
import random
import time
```

```python
user_m = int(input()) #строки
user_n = int(input()) #столбцы
user_min_limit = int(input())
user_max_limit = int(input())

#user_m, user_n, user_min_limit, user_max_limit
#создание матрицы
mas = []
for i in range(user_m):
   mas.append([0]*user_n) 

#заполнение матрицы
for i in range (user_m):
    for j in range (user_n):
        mas[i][j] = random.randint(user_min_limit, user_max_limit)

#вывод матрицы
for i in range (user_m):
    for j in range (user_n):
        print (mas[i][j], end = ' ')
    print()
```

### Задание №3
Реализовать методы сортировки строк числовой матрицы в соответствии с

```python
import copy
```

```python
# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.
start_time = time.time()
#TODO

#вывод исходной матрицы        
for i in range (user_m):
    for j in range (user_n):
        print (mas[i][j], end = ' ')
    print()
print()

#сортриовка
mas_1 = copy.deepcopy(mas)

def selection_sort(mas_1):
    min_ind = 0
    while min_ind < user_n-1:
        qur_min = min_ind
        for j in range (min_ind + 1, user_n):
            if mas_1[j] < mas_1[qur_min]:
                qur_min = j
        temp = mas_1[min_ind]
        mas_1[min_ind] = mas_1[qur_min]
        mas_1[qur_min] = temp
        min_ind +=1
    return mas_1

        
#вывод отсортированной матрицы        
for i in range (user_m):
    mas_1[i] = selection_sort(mas_1[i])
    for j in range (user_n):
        print (mas_1[i][j], end = ' ')
    print()

time_1 = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(time_1))
```

```python
# Сортировка вставкой.
start_time = time.time()
#TODO

#вывод исходной матрицы        
for i in range (user_m):
    for j in range (user_n):
        print (mas[i][j], end = ' ')
    print()
print()

#сортриовка
mas_2 = copy.deepcopy(mas)

def insertion_sort(mas_2):
    for sorted_ind in range (1, user_n):
        temp = mas_2[sorted_ind]
        j = sorted_ind - 1
        while j >= 0 and temp < mas_2[j]:
            mas_2[j+1] = mas_2[j]
            j -= 1
        mas_2[j+1] = temp
    return mas_2

#вывод отсортированной матрицы        
for i in range (user_m):
    mas_2[i] = insertion_sort(mas_2[i])
    for j in range (user_n):
        print (mas_2[i][j], end = ' ')
    print()

time_2 = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(time_2))
```

```python
# Сортировка Шелла.
start_time = time.time()
#TODO

#вывод исходной матрицы        
for i in range (user_m):
    for j in range (user_n):
        print (mas[i][j], end = ' ')
    print()
print()

#сортриовка
mas_4 = copy.deepcopy(mas)
def shell_sort(mas_4):
    gap = user_n //2
    while gap > 0:
        for insert_ind in range (gap, user_n):
            temp = mas_4[insert_ind]
            j = insert_ind
            while j >= gap and mas_4[j - gap] > temp:
                mas_4[j] = mas_4[j - gap]
                j -= gap
                mas_4[j] = temp
        gap //= 2
    return mas_4


#вывод отсортированной матрицы        
for i in range (user_m):
    mas_4[i] = shell_sort(mas_4[i])
    for j in range (user_n):
        print (mas_4[i][j], end = ' ')
    print()

time_4 = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(time_4))
```

```python
# Турнирная сортировка.
start_time = time.time()
#TODO

#вывод исходной матрицы        
for i in range (user_m):
    for j in range (user_n):
        print (mas[i][j], end = ' ')
    print()
print()

mas_6 = copy.deepcopy(mas)
class PriorityHeap:
    def __init__(self) -> None:
        self.cache: list[int] = []
        self.size: int = 0
        self.max_size = 32

    def pop_max(self) -> any: ###
        res = self.cache[0]
        self.size -= 1
        self.cache[0] = self.cache[self.size]
        self.cache.pop(self.size)
        self.iter_heapify(0)
        return res

    def max_heap_insert(self, key: int):
        self.cache.append(-1000000000000000)
        self.insert_key(key, self.size)       #iter_key
        self.size += 1
        
    def insert_key(self, value: int, i: int):
        self.cache[i]  = value
        parent = self.parent(i)
        while i > 0 and self.cache[parent] > self.cache[i]:
            self.cache[parent], self.cache[i] = self.cache[i], self.cache[parent]
            i = parent
            parent = self.parent(i)

    def parent(self, i: int) -> int:
        res = (i >> 1) - 1
        if res < 0:
            return 0
        else:
            return res

    def iter_heapify(self, i: int) -> None:
        while True:
            l: int = (i << 1) + 1
            r: int = (i << 1) + 2
            largest: int = i

            if l < self.size and self.cache[l] < self.cache[largest]:
                largest = l

            if r < self.size and self.cache[r] < self.cache[largest]:
                largest = r

            if largest ==i:
                break
            else:
                self.cache[i], self.cache[largest] = self.cache[largest], self.cache[i]
                i = largest

                
                
def tournament_sort(a: list[int]) -> list[int]:
    a_len: int = len(a)
    heap = PriorityHeap()
    winners: list[int] = []
    losers: list[int] = []

    while heap.size < heap.max_size and heap.size < a_len:
        elem = a[heap.size]
        if winners and winners[-1] > elem:
            losers.append(elem)
        else:
            heap.max_heap_insert(a[heap.size])

    for i in range(heap.size, a_len):
        elem: int = a[i]

        if winners and winners[-1] > elem:
            losers.append(elem)
        else: 
            max_elem: int = heap.pop_max()

            if not winners:
                winners.append(max_elem)
            elif winners[-1] <= max_elem:
                winners.append(max_elem)
            else:
                losers.append(max_elem)

            heap.max_heap_insert(elem)
    
    while heap.size > 0:
        max_elem: int = heap.pop_max()

        if not winners:
            winners.append(max_elem)
        elif winners[-1] <= max_elem:
            winners.append(max_elem)
        else: 
            losers.append(max_elem)
    
    if not losers:
        return winners
    else:
        winners2: list[int] = tournament_sort(losers)

        i: int = 0
        j: int = 0
        winners1_len: int = len(winners)
        winners2_len: int = len(winners2)

        winners_res: list[int] = []

        while i < winners1_len and j < winners2_len:
            if winners[i] < winners2[j]:
                winners_res.append(winners[i])
                i += 1
            else:
                winners_res.append(winners2[j])
                j += 1
        while i < winners1_len:
            winners_res.append(winners[i])
            i += 1
        while j < winners2_len:
            winners_res.append(winners2[j])
            j += 1
        return winners_res


#вывод отсортированной матрицы        
for i in range (user_m):
    mas_6[i] = tournament_sort(mas_6[i])
    for j in range (user_n):
        print (mas_6[i][j], end = ' ')
    print()

time_6 = round((time.time() - start_time)*1000)
print("--- {0} ms ---".format(time_6))
```
